```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'font.size': 12})

# Funktion zur Berechnung der Zufallsgröße Xn
def calculate_Xn(n):
    outcomes = np.arange(2**n)
    probabilities = [1 / (2**n)] * (2**n)
    Xn_values = [sum([((outcome >> i) & 1) / 2**(i+1) for i in range(n)]) for outcome in outcomes]
    return Xn_values, probabilities


# Setze das Layout der Subplots (3 Spalten, 3 Zeilen)
fig, axes = plt.subplots(3, 3, figsize=(10, 10))
fig.tight_layout(pad=5.0)

# Erstelle die Graphen für n=1 bis n=9
for n in range(1, 10):
    row = (n-1) // 3  # Berechnung der Zeile
    col = (n-1) % 3   # Berechnung der Spalte
    ax = axes[row, col]  # Zugriff auf den entsprechenden Subplot
    
    Xn_values, probabilities = calculate_Xn(n)
    
    # Erzeuge den Stem-Plot ohne Marker und mit dickerem Stab
    stem_container = ax.stem(Xn_values, probabilities, basefmt=" ", 
                             linefmt='b-',  # grüne Linien
                             markerfmt='r.',  # rote Punkt
                             use_line_collection=True)
    
    # Ändere die Stabdicke direkt über den Rückgabewert von stem()
    plt.setp(stem_container.stemlines, 'linewidth',1)  # Setze die Stabdicke
    
    ax.set_title(f' $n = {n}$')
    ax.set_xlabel(r'$k$')  # X-Achsenbeschriftung mit LaTeX
    ax.set_ylabel(r'$\mathbb{P}(X_n = k)$')  # Y-Achsenbeschriftung mit LaTeX
    plt.tight_layout()
    #plt.savefig('Graphen-3x3.png') # speichern der Grafik als png-Datei: # entfernen
plt.show()
```
